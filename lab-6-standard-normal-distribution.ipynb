# Histogram

```python
# histogram plot
plt.hist(
    df['Current'], bins=5, 
    alpha = 0.8,
    color='white',
    edgecolor= 'red'
)

plt.title(
    'Histogram of Current Values',
    
)
"""
plt.plot(
    x,
    line,
    color='blue'
)
"""

plt.xlabel('Current (mA)')
plt.ylabel('Frequency')
plt.show()
```

```python
#normal distribution curve
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

#x-axis
x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)



plt.plot(
    x, p,
    color='violet',
    linewidth=2
)
plt.title(
    'Normal Distribution of Current Values',
)

plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
plt.show
```

```python
plt.hist(
    df['Current'], bins=5,
    density= True,
    alpha = 0.8,
    color='green',
    edgecolor= 'red'
)
#normal distribution curve
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

#x-axis
x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)



plt.plot(
    x, p,
    color='violet',
    linewidth=2
)
plt.title(
    'Normal Distribution of Current Values',
)

plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
plt.show
```

# Standard Normal Distribution

```python
#normal distribution curve
mean = df['Z-score'].mean()
std = df['Z-score'].std(ddof=1)

#x-axis
x_min = df['Z-score'].min()
x_max = df['Z-score'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(
    x, p,
    color='violet',
    linewidth=2
)
plt.title(
    'Normal Distribution of Current Values N(0,1)',
)

plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.show
```
