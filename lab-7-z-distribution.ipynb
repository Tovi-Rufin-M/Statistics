```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from scipy import stats
```

```python
# load dataset                                                  to sort in sample
df = pd.read_csv(r'dataset\transformer-voltage.csv', delimiter=",", index_col='Sample')
df
```

```python
#summary of statistics
df.info()
```

```python
df.describe()
```

```python
alpha = 0.05
z_critical = stats.norm.ppf(1-alpha/2)
print(z_critical)
```

```python
pop_std = 5

ci_lower, ci_upper = stats.norm.interval(
    confidence=0.95,
    loc = df['Voltage'].mean(),
    scale = pop_std/np.sqrt(df['Voltage'].count())
)

print(f"CI: [{ci_lower:.2f},{ci_upper:.2f}]")
```

```python
#Voltage distribution 
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

#x-axis
x_min = df['Voltage'].min()
x_max = df['Voltage'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.hist(
    df['Voltage'], bins=5,
    density= True,
    alpha = 0.8,
    color='green',
    edgecolor= 'red'
)

plt.plot(
    x, p,
    color='violet',
    linewidth=2
)
plt.title(
    'Distribution of Voltage Values',
)

plt.xlabel('Voltage (V)')
plt.ylabel('Probability Density')

plt.axvline(
    x=ci_lower,
    color = 'red',
    linestyle = "--"
)
plt.axvline(
    x=ci_upper,
    color = 'red',
    linestyle = "--"
)
plt.show()
```
