```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from scipy import stats
```

```python
# load dataset                                                  to sort in sample
df = pd.read_csv(r'dataset\transformer-voltage-10-sample.csv', delimiter=",", index_col='Sample')
df
```

```python
#summary of statistics
df.info()
```

```python
df.describe()
```

```python
alpha = 0.05
n = df['Voltage'].count()

t_critical = stats.t.ppf(1-alpha/2,n-1)
print(t_critical)
```

```python
std = df['Voltage'].std(ddof=1)

ci_lower, ci_upper = stats.t.interval(
    confidence=0.95,
    df = n-1,
    loc = df['Voltage'].mean(),
    scale = std/np.sqrt(df['Voltage'].count())
)

print(f"CI: [{ci_lower:.2f} to {ci_upper:.2f}]")
```

```python
#Voltage distribution 
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

#x-axis
x_min = df['Voltage'].min()
x_max = df['Voltage'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.hist(
    df['Voltage'], bins=5,
    density= True,
    alpha = 0.8,
    color='green',
    edgecolor= 'red'
)

plt.plot(
    x, p,
    color='orange',
    linewidth=2
)
plt.title(
    'Distribution of Voltage Values',
)

plt.xlabel('Voltage (V)')
plt.ylabel('Probability Density')

plt.axvline(
    x=ci_lower,
    color = 'red',
    linestyle = "--"
)
plt.axvline(
    x=ci_upper,
    color = 'red',
    linestyle = "--"
)
plt.show()
```

# Show Inventory

```python
# load dataset                                                  to sort in sample
df = pd.read_csv(r'dataset\shoe-inventory.csv', delimiter=",")
df
```

```python
df.shape
```

```python
mean_size_6 = df.iloc[0,1:13].mean()
print(mean_size_6)
```

```python
std_size_6 = df.iloc[0,1:13].std()
std_size_6
```

```python
# size 6 95% CI
n = df.shape[1]-1 #returns 13 to 12
n
lower_6, upper_6 = stats.t.interval(
    confidence=0.95,
    df = n-1,
    loc = mean_size_6,
    scale = std_size_6/np.sqrt(n)
)
print(f"CI_6: [{lower_6:.2f} to {upper_6:.2f}]")
```

```python
mean = df.iloc[0:17,1:13].mean(axis=1)
std = df.iloc[0:17,1:13].std(axis=1)

df['ci_lower'], df['ci_upper'] = stats.t.interval(
    confidence=0.95,
    df = n-1,
    loc = mean,
    scale = std/np.sqrt(n)
)
```

```python
df
```

```python
df['ci_lower'] = df['ci_lower'].fillna(0)
df['ci_upper'] = df['ci_upper'].fillna(0)
```

```python
df['ci_lower'] = df['ci_lower'].astype(int).round()
df['ci_upper'] = df['ci_upper'].astype(int).round()
```
