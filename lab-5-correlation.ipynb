```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from scipy import stats
```

```python
df = pd.read_csv(r'dataset\current-power.csv', delimiter=",")

df
```

```python
df.describe()
```

```python
# correlation config of current and power
df.corr()
```

```python
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current VS Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.show()
```

```python
# regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt. figure(figsize=(7,4))
plt.scatter(
    x, y, 
    color='#1c1f24',
    alpha=0.6,
    label='Data points'
)
# line plot
plt.plot(
    x,
    line,
    color='blue',
    label=f'Regression line (r = {r_value:.3f})'
)
plt.title('Device Current VS Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.legend()
plt.show()
```

## Air Quality

```python
# Load Dataset
air = pd.read_csv(r"dataset\air-quality.csv", delimiter=",")
air
```

### NOx

```python
x = air['NOx']
y = air['Relative_Humidity']

# regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt. figure(figsize=(7,4))
plt.scatter(
    x, y, 
    color="#68b7cf",
    alpha=0.2,
    label='Data points'
)
# line plot
plt.plot(
    x,
    line,
    color='violet',
    label=f'Regression line (r = {r_value:.3f})'
)
plt.title('NOx VS Humidity Correlation')
plt.ylabel('Relative_Humidity (%)')
plt.xlabel('NOx (ug/cu.m)')
plt.legend()
plt.show()
```
